```python
from sklearn.preprocessing import FunctionTransformer

# 1. Define a helper function to join multiple columns into one string
def combine_text_columns(X):
    # X is the 2D array from the imputer. 
    # We convert to DataFrame to use easy string joining
    return pd.DataFrame(X).astype(str).agg(' '.join, axis=1)

# 2. Update the pipeline
text_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='')),
    
    # NEW STEP: Join the columns together
    ('combiner', FunctionTransformer(combine_text_columns, validate=False)),
    
    ('tfidf', TfidfVectorizer(
        lowercase=True,
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.95,
        sublinear_tf=True
    ))
])
```

```python
# Pipeline for boolean columns
bool_pipeline = Pipeline(steps=[
    # Handle missing values by replacing them with the most frequent value
    ('imputer', SimpleImputer(strategy='most_frequent')),
    
    # Convert boolean values to integers (True -> 1, False -> 0)
    ('to_int', FunctionTransformer(lambda x: x.astype(int), validate=False))
])
```
